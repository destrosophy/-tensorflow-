```python
%matplotlib inline
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt 
import tensorflow.examples.tutorials.mnist.input_data as input_data
mnist=input_data.read_data_sets("MNIST_data/",one_hot=True)
```

```python
HH=256
x1=tf.placeholder("float64",[None,28*28],name="x1")
y1=tf.placeholder("float64",[None,10],name="y1")
w1=tf.Variable(np.random.randn(784,HH),dtype="float64")
b1=tf.Variable(np.zeros(HH),dtype="float64")
pre1=tf.nn.relu(tf.matmul(x1,w1)+b1)
w2=tf.Variable(np.random.randn(HH,10),dtype="float64")
b2=tf.Variable(np.zeros(10),dtype="float64")
pre2=tf.matmul(pre1,w2)+b2
predict=tf.nn.softmax(pre2)
leanning_rate=0.05
train_epoches=100
batch_size=50
batch_total=int(len(mnist.train.images)/batch_size)
loss_function=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y1,logits=pre2))
optimizer=tf.train.GradientDescentOptimizer(leanning_rate).minimize(loss_function)
correct=tf.equal(tf.argmax(predict,1),tf.argmax(y1,1))
accuracy=tf.reduce_mean(tf.cast(correct,tf.float32))
init=tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    for t in range(train_epoches):
        for i in range(batch_total):
            xs,ys=mnist.train.next_batch(batch_size)
            _,loss,acc=sess.run([optimizer,loss_function,accuracy],feed_dict={x1:xs,y1:ys})
        print(acc)
```

                           前馈神经网络（FNN）
    组成：前馈神经包括大量神经元，以层组织：一个输入层，一个或多个隐藏层，一个输出层，每个神经元都与前一层的所有神经元连接，并非所有连接都相同，它们具有不同的权值，这些连接的权值承载了整个网络的信息
    
    意义：卷积神经网络、循环神经网络等都是该网络的变种
    
    流程：数据进入输入层然后逐层通过网络，最终达到输出层，在此过程中，各层之间没有相互反馈
    
    隐藏层层数选择问题：如果一个神经网络由很少的隐藏层或者神经元组成，那么它对未知函数的拟合精度可能不会太高，如果包含大量隐藏层，我们可能会遇到过拟合的问题，网络的泛化能力也会变差
    
    权重和偏差：每个单元i由一个输入向量x（x1，x2...xn）和一个权重向量w（wi1,wi2...win）定义，神经网络元i的值为输入的加权个：net=∑wijxj，这些权重里有一个特殊权重，成为偏差bias，net=∑wijxj+bi
    
    传递函数：Sigmoid，Tanh,Softmax,Rulu=max(0,x)等具体表达式可以百度，为了使每个神经元的输出值都落在0和1之间，如果不添加传递函数，每一步过程都相当于线性变换，使用非线性激活函数，我们便能够从输入输出之间生成非线性映射。
    
    softmax分类器：数据必须使用One-Hot独热编码，如果是二分类问题那么它对应的独热编码可能是[0,1]，只有一个值为1其他值都为0，一般在前馈神经网络的最后悔使用softmax+交叉熵定义损失函数
    
    误差反向传播算法BP：（复合函数求导问题，责任追究问题）
